```python
import random
from typing import List
import matplotlib.pyplot as plt
```

```python
def randomWalk(x, t, k, n):
  walks = [x]
  t = [t]
  t_change = 1
  for i in range(n):
    num = random.random()
    if num < 0.5:
      walks.append(walks[-1] - k)
    else:
      walks.append(walks[-1] + k)

    t.append(t[-1] + t_change)

  return walks, t
```

```python
def plotWalk():
  walks, t = randomWalk(5, 0, 2, 1000)
  print(t)
  plt.plot(walks)
  plt.show()
```

```python
plotWalk()
```

```python
def randomWalk_2D(x, y, t, k, n, inc_time = 1):
  walks = [(x, y)]
  time = [t]
  for i in range(n):
    num = random.random()
    if 0.0 < num < 0.25:
      walks.append((walks[-1][0], walks[-1][1] + k))
    elif 0.25 <= num < 0.5:
      walks.append((walks[-1][0] + 1, walks[-1][1]))
    elif 0.5 <= num < 0.75:
      walks.append((walks[-1][0], walks[-1][1] - 1))
    else:
      walks.append((walks[-1][0] - k, walks[-1][1]))

    time.append(time[-1] + inc_time)

  return walks, time
```

```python
#plot 2d walks

walks, t = randomWalk_2D(10, 5, 0, 1, 1000)
x, y = zip(*walks)
plt.plot(x, y)
plt.show()
```
